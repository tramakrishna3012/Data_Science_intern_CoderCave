```python
#drop null values
df.dropna(inplace = True)
```

```python
#merge Star1,Star2,Star3,Star4 in new column cast
df['Cast'] = df['Star1'].astype(object) + ',' + df['Star2'].astype(object) + ',' 
+ df['Star3'].astype(object) + ',' + df['Star4'].astype(object)
#drop Star1,Star2,Star3,Star4 columns 
df = df.drop(columns = ['Star1' ,'Star2','Star3','Star4'])
```

```python
df.columns
```

```python
#Reindexing all columns
df = df.reindex(['Poster_Link', 'Series_Title', 'Released_Year', 'Certificate',
       'Runtime', 'Genre', 'IMDB_Rating', 'Overview', 'Meta_score', 'Director',
       'Cast','No_of_Votes', 'Gross'] , axis = 1)
```

```python
df.columns
```

```python
#Sort the movies by Released Year
df.sort_values('Released_Year')
```

```python
#Filter the and select only movies between 2000 to 2023
#Define the start and end year for filtering
start_year = '2000'
end_year = '2023'

#Filter the data based on the Released year column
filtered_df = df[df['Released_Year'].str.isnumeric() & (df['Released_Year'].astype(str) >= str(start_year)) 
                 & (df['Released_Year'].astype(str) <= str(end_year))]
filtered_df
```

```python
#Again Sort the filtered values
df_sort = filtered_df.sort_values('Released_Year')
df_sort
```

```python
#Sort the above sorted by there IMDB Rating , Mete Score , No. of Votes
df_rec = df_sort.sort_values(['IMDB_Rating' , 'Meta_score' , 'No_of_Votes'] , ascending = [False,False,False])
df_rec
```

```python
#Top 50 movies from above filtered movie list
top_50_movies = df_rec.head(50)
top_50_movies
```

```python
#Group top 50 movies by genres and sort
movies_genre = top_50_movies.groupby('Genre')['Series_Title'].count().sort_values(ascending = False)
```

```python
movies_genre = pd.DataFrame(movies_genre).rename(columns = {'Series_Title':'Count'})
movies_genre
```

```python
#Plot graph between Genre and No. of Movies
movies_genre.head(12).plot(kind = 'bar')
plt.xlabel("Genre")
plt.ylabel("No. of Movies")
plt.title("The most popular genre of movies")
plt.show()
```

```python
#Now Group movies by directors and top 20 directors
movies_directors = top_50_movies.groupby('Director')['Series_Title'].count().sort_values(ascending = False)
movies_directors = pd.DataFrame(movies_directors).rename(columns = {'Series_Title' : 'Count'})
movies_directors.head(20)
```

```python
top_20 = top_50_movies.sort_values(by = ['IMDB_Rating']).tail(20) # top 20 high rated movies
top_20
```

```python
#Plot graph between Top rated movies and directors
Directors = top_20['Director']
IMDB_Rating = top_20['IMDB_Rating'].sort_values(ascending = False)
fig = plt.figure(figsize = (20,10))
plt.xlabel("Directors")
plt.ylabel("IMDB_Rating")
plt.title("The most top-rated movies")
plt.bar(Directors[0:10],IMDB_Rating[0:10])
plt.show()
```
